## 2021 Day 1, Puzzle 1

```python
# Read the data and choose the data structure
data_file = 'data/input.txt'
with open(data_file, 'r') as input_file:
    lines = input_file.read().splitlines()
```

```python
lines
```

```python
# Redefine lines
lines = [int(line) for line in lines]
```

```python
lines
```

```python
# Count starts in 0 and will increase every time a line value increases
count_increased = 0
```

```python
#Start a loop from index 1 through the lines
for i in range(1, len(lines)):
    if lines[i] > lines[i - 1]:
        count_increased += 1
```

```python
# Print the result for Puzzle 1 
print(f'Puzzle 1: {count_increased}')
```

## 2021 Day 1, Puzzle 2

```python
lines
```

```python
# Count starts in 0 and will increase every time a line value increases
count_increased = 0
```

```python
# Start a loop from index 3 through the lines
for a in range(3, len(lines)):
    # Total of the actual window
    actual_total = lines[a-2] + lines[a-1] + lines[a]
    # Total of the prior window
    prior_total = lines[a-3] + lines[a-2] + lines[a-1]
    # Compare the totals and increase if the actual sum is larger
    if actual_total > prior_total:
        count_increased += 1
```

```python
# Print the result for Puzzle 2
print(f'Puzzle 2: {count_increased}')
```
